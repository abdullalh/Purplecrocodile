```python
class Customer:
    def __init__(self, customerID, name, email, contactNumber, deliveryAddress):
        self.customerID = customerID
        self.name = name
        self.email = email
        self.contactNumber = contactNumber
        self.deliveryAddress = deliveryAddress

    def displayCustomerInfo(self):
        print(f"Customer ID: {self.customerID}\nName: {self.name}\nEmail: {self.email}\n"
              f"Contact: {self.contactNumber}\nDelivery Address: {self.deliveryAddress}\n")


class Item:
    def __init__(self, itemCode, description, quantity, unitPrice):
        self.itemCode = itemCode
        self.description = description
        self.quantity = quantity
        self.unitPrice = unitPrice
        self.totalPrice = quantity * unitPrice

    def displayItemInfo(self):
        print(f"Item Code: {self.itemCode} | Description: {self.description} | Quantity: {self.quantity} | "
              f"Unit Price: AED {self.unitPrice} | Total Price: AED {self.totalPrice}")


class Order:
    def __init__(self, orderID, orderDate, totalPrice, status, paymentMethod):
        self.orderID = orderID
        self.orderDate = orderDate
        self.totalPrice = totalPrice
        self.status = status
        self.paymentMethod = paymentMethod

    def displayOrderDetails(self):
        print(f"Order ID: {self.orderID} | Order Date: {self.orderDate}\nTotal Price: AED {self.totalPrice} | "
              f"Status: {self.status} | Payment Method: {self.paymentMethod}\n")


class DeliveryNote:
    def __init__(self, noteID, recipientName, deliveryAddress, totalWeight, deliveryDate):
        self.noteID = noteID
        self.recipientName = recipientName
        self.deliveryAddress = deliveryAddress
        self.totalWeight = totalWeight
        self.deliveryDate = deliveryDate

    def displayNoteDetails(self):
        print(f"Delivery Note ID: {self.noteID}\n"
              f"Recipient: {self.recipientName}\n"
              f"Delivery Address: {self.deliveryAddress}\n"
              f"Total Weight: {self.totalWeight} kg\n"
              f"Delivery Date: {self.deliveryDate}\n")


class Invoice:
    def __init__(self, invoiceID, subtotal, taxes, totalAmount, dueDate, paymentStatus):
        self.invoiceID = invoiceID
        self.subtotal = subtotal
        self.taxes = taxes
        self.totalAmount = totalAmount
        self.dueDate = dueDate
        self.paymentStatus = paymentStatus

    def displayInvoiceDetails(self):
        print(f"Invoice ID: {self.invoiceID}\n"
              f"Subtotal: AED {self.subtotal}\n"
              f"Taxes and Fees: AED {self.taxes}\n"
              f"Total Charges: AED {self.totalAmount}\n"
              f"Due Date: {self.dueDate}\n"
              f"Payment Status: {self.paymentStatus}\n")


# Create Two Objects for Each Class
customer1 = Customer("CUST001", "Johnson", "johnson@example.com", "0501234567", "Dubai, UAE")
customer2 = Customer("CUST002", "Alice", "alice@example.com", "0559876543", "Abu Dhabi, UAE")

item1 = Item("ITEM123", "Laptop", 2, 3500.00)
item2 = Item("ITEM456", "Smartphone", 1, 2800.00)

order1 = Order("ORD001", "2025-01-25", item1.totalPrice, "Processing", "Credit Card")
order2 = Order("ORD002", "2025-02-10", item2.totalPrice, "Shipped", "Cash on Delivery")

deliveryNote1 = DeliveryNote("DEL001", "Johnson", "Dubai, UAE", 5, "2025-01-28")
deliveryNote2 = DeliveryNote("DEL002", "Alice", "Abu Dhabi, UAE", 3, "2025-02-12")

invoice1 = Invoice("INV001", item1.totalPrice, 150.00, item1.totalPrice + 150.00, "2025-02-05", "Unpaid")
invoice2 = Invoice("INV002", item2.totalPrice, 100.00, item2.totalPrice + 100.00, "2025-02-15", "Paid")

# Display details in sequence: Delivery Note -> Customer -> Order -> Item -> Invoice
print("============== CUSTOMER 1 ==============")
deliveryNote1.displayNoteDetails()
customer1.displayCustomerInfo()
order1.displayOrderDetails()
item1.displayItemInfo()
invoice1.displayInvoiceDetails()

print("============== CUSTOMER 2 ==============")
deliveryNote2.displayNoteDetails()
customer2.displayCustomerInfo()
order2.displayOrderDetails()
item2.displayItemInfo()
invoice2.displayInvoiceDetails()
```
